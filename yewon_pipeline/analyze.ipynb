# Excluding incorrect mask

```python
import pandas as pd
pd.set_option('display.max_rows', 100) 

df = pd.read_csv('/local-ssd/yl3427/test/runs/eval/det_cls_grid.csv')
df['classifier'] = df['checkpoint'].apply(lambda x: x.split('/')[-2].replace('cropped_', ''))
df.drop(columns=['checkpoint', 'data_root', 'iou_thresh', 'total_gt_faces', 'images', 'matched_detections'], inplace=True)
df.head(20)
df['classifier'].unique()
```

```python
df[['detector', 'det_recall', 'classifier', 'cls_acc', 'cls_samples']].groupby(['detector',  'classifier']).mean()
```

```python
df[df['classifier'].str.startswith('vit')].sort_values(by=['det_recall', 'cls_acc'], ascending=False)
```

```python
df.groupby(['detector', 'classifier']).mean().reset_index()
```
